# Hourly demand (overall) 

<img src="../outputs/figures/Trip_demand_by_hour_chart.png" style="width:800px;height:500px"/>

Insight:
Ridership shows three clear demand windows: late night (1–3 AM), midday (11 AM–1 PM), and evening (5–8 PM).
Despite the unexpected late-night peak, average trip duration remains stable (14–15 minutes), 
indicating consistent rider behavior throughout the day.

Implication:
The system supports both commuter and leisure usage, suggesting demand is not limited to traditional work travel.

# Weekday vs Weekend hourly comparison

<img src="../outputs/figures/Trip_demand_by_hour_weekend_vs_weekdaychart.png" style="width:800px;height:500px"/>


Insight:
Overall ridership is driven by weekdays; however, late-night demand (12–7 AM) is concentrated on weekends, indicating strong leisure and nightlife usage.

Key Pattern:

Weekdays → stable daytime demand with mild lunch peak

Weekends → sharp spikes at 1–3 AM and 6–8 PM

Implication:
The system appears more lifestyle-oriented than commuter heavy, especially outside weekday hours.

# Trips per month (seasonality)

<img src="../outputs/figures/Trip_demand_per_month_chart.png" style="width:800px;height:500px"/>


Insight:
Ridership follows a clear seasonal pattern, with peaks in February–March and August–October, and declines in January and late-year months.

Implication:
Demand is likely influenced by seasonal factors such as weather, indicating opportunities for seasonal operational planning.

# Subscriber vs Casual trip count Subscriber vs Casual avg duration

<img src="../outputs/figures/trip_count_per_usertype_stats_table.png" style="width:; height:"/>


Insight:
Subscribers generate the majority of trips, confirming recurring users drive demand. However, trip duration and distance are consistent across both rider types.

Implication:
While usage frequency differs, riding behavior is similar suggesting marketing strategies should focus on increasing ride frequency rather than altering ride experience.

# Temperature vs avg rides per day


<img  src="../outputs/figures/avg_and_ride_count_per_temp_range.png" style="width: ; height:"/>

Ridership peaks in moderate temperatures (15–25°C) and declines sharply in extreme cold (<10°C) or heat (>30°C).

Implication:
Demand is highly temperature sensitive, with optimal operational capacity required during mild weather conditions.

# Rain category vs total trips

<img  src="../outputs/figures/trip_count_per_rain_cat_temp_range.png" style="width: ; height:"/>


Insight:
Mild temperatures drive ridership more strongly than the absence of rain. Heavy rain significantly suppresses demand, especially among casual riders.

Implication:
Weather resistant infrastructure or promotions during mild rainy conditions could help stabilize usage.
